```python
df['Total_Revenues']=df['TotalPrice'].sum()
print(df['Total_Revenues'])
Total_Revenues
```

```python
top_products=df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(3)
print(top_products)
```

```python
category_renues=df.groupby("Category")['TotalPrice'].sum().sort_values(ascending=False)
print(category_renues)
```

```python
sales_by_region=df.groupby("Region")["TotalPrice"].sum().sort_values(ascending=False)
print(sales_by_region)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualization 1  Top 5 Products by Revenue
top_5 = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
sns.barplot(x=top_5.values, y=top_5.index)
plt.title("Top 5 Products by Revenue")
plt.xlabel("Revenue ()")
plt.ylabel("Product")
plt.show()
```

```python
# Line chart - Sales over Time
plt.figure(figsize=(10,5))
df.groupby('Date')['TotalPrice'].sum().plot(marker='o', color='orange')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram - Total Price Distribution
plt.figure(figsize=(8,5))
plt.hist(df['TotalPrice'], bins=10, color='purple', edgecolor='black')
plt.title("Distribution of Total Sales per Order")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot - Quantity vs TotalPrice
plt.figure(figsize=(8,5))
sns.scatterplot(x='Quantity', y='TotalPrice', hue='Category', data=df, palette='Set2')
plt.title("Quantity vs Total Sales")
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart - Sales Share by Region
plt.figure(figsize=(6,6))
region_sales = df.groupby('Region')['TotalPrice'].sum()
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Sales Share by Region")
plt.tight_layout()
plt.show()
```
